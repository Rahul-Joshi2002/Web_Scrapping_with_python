```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
Record = []
ODI_Record_Holder = []
TEST_Record_Holder = []
T20_Record_Holder = []
data = pd.DataFrame()
```

```python
url = "https://stats.espncricinfo.com/ci/content/records/283792.html"
page = requests.get(url)
htmlContent = page.content
```

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

```python
body = soup.find('tbody')
body
```

```python
caption = soup.find('table', class_="engineTable").find('caption').text
print(caption)
```

```python
entry = body.find('tr', class_="data1").find('td').text
print(entry)
```

```python
Record.append(caption)
ODI_Record_Holder.append(entry)
```
