```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
Record = []
Record_Holder = []
data = pd.DataFrame()
```

```python
url = "https://stats.espncricinfo.com/ci/content/records/223375.html"
page = requests.get(url)
htmlContent = page.content
```

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

```python
body = soup.find('tbody')
body
```

```python
caption = soup.find('table', class_="engineTable").find('caption').text
print(caption)
```

```python
entry = body.find('tr', class_="data1").find('td').text
print(entry)
```

```python
Record.append(caption)
Record_Holder.append(entry)
```

```python
caption = soup.find('div', class_="icc-home").text
i = 0;
s = ""
for element in caption:
    if(i > 85):
        s = s + element
    i = i + 1
s = s.replace('\n', '').replace('\t' , '').strip(" ")
print(caption)
print(s)
```
