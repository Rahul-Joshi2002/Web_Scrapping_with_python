```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://stats.espncricinfo.com/ci/engine/player/34102.html?class=2;template=results;type=allround"
page = requests.get(url)
htmlContent = page.content
```

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

```python
tables = soup.find_all('table', class_="engineTable")
```

```python
for item in tables:
	print(item.get_text().strip())
```

```python
x = soup.find('tr', class_="head")
```

```python
para = []
for item in x:
    item = item.get_text().replace('\n', ' ').strip(" ")
    para.append(item)
while("" in para) :
    para.remove("")
```

```python
para
```

```python
y = soup.find('tr', class_="data1")
```

```python
stat = []
i = 0
for item in y:
    if i == 28:
        break
    item = item.get_text().replace('\n', ' ').strip(" ")
    stat.append(item)
    i = i + 1
stat.remove("overall")
while("" in stat) :
    stat.remove("")
```

```python
stat
```

```python
data = pd.DataFrame()
data['Parameters'] = para
data['Statistics'] = stat
data
```

```python
data.to_csv("Rohit (ODI).csv", index = False)
```
