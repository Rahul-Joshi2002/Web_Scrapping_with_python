```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://stats.espncricinfo.com/ci/content/records/283193.html"
page = requests.get(url)
htmlContent = page.content
```

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

```python
tables = soup.find_all('table', class_="engineTable")
```

```python
caption = soup.find('caption')
print(caption.text)
```

```python
head = soup.find('tr', class_="head")
head
```

```python
headings = []
for heading in head:
    heading = heading.text.strip("").replace('\n', "");
    if(heading == ''):
        continue
    headings.append(heading)
headings
```

```python
items = soup.find('table', class_="engineTable").find('tbody')
items
```

```python
stat = []
for item in items:
    item = item.get_text().replace('\n', ' ').strip(" ")
    if(item == ''):
        continue
    stat.append(item)
```

```python
stat
```

```python
threads = soup.find_all('tr', class_="data1")
threads
```

```python
data = pd.DataFrame()
data['Headings'] = headings
```

```python
i = 0
for thread in threads:
    dirs = thread.find_all('td')
    info = []
    for dir in dirs:
        dir_info = dir.text.replace('\n', '').strip("")
        info.append(dir_info)
    print(info)
    i = i + 1
    data[i] = info
```

```python
data
```

```python
data.to_csv('Bowling_Records_(ODI).csv', index = False)
```
